```python
x =  np.linspace(-np.pi, np.pi, 100)
```

```python
plt.plot(x, np.sin(x), x, np.cos(x))
```

```python
plt.quiver(0, 0, 1, 2, color='r', angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, 4, 3, color='b', angles='xy', scale_units='xy', scale=1)
plt.quiver(0, 0, 5, 5, color='g', angles='xy', scale_units='xy', scale=1)
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.show()
```

```python
x = np.linspace(20, 30)
y = np.linspace(0, 20)
X,Y = np.meshgrid(x, y)
z = -X + 2 * Y
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z)
plt.xlabel("$x_1$")
plt.xlabel("$x_2$")
plt.title("The plane is $x_1-2x_2+x_3=0$")
plt.show()
```
