<p ALIGN="left"><b><font size="5">Naming Coordination Compounds</font></b></p>
<p ALIGN="left">A <b>complex</b> is a substance in which a metal atom or ion is 
  associated with a group of neutral molecules or anions called <b>ligands</b>. 
  <b>Coordination compounds</b> are neutral substances (i.e. uncharged) in which 
  <b>at least</b> one ion is present as a complex. You will learn more about coordination 
  compounds in the lab lectures of experiment 4 in this course. </p>
<p ALIGN="left">The coordination compounds are named in the following way. (At 
  the end of this tutorial we have some examples to show you how coordination 
  compounds are named.)</p>
  <blockquote>
  <p ALIGN="left"><b>A.</b> To name a coordination compound, no matter whether the 
    complex ion is the cation or the anion, <b>always name the cation before the 
    anion</b>. (This is just like naming an ionic compound.)</p>
      
  <p ALIGN="left"><b>B</b>. In naming the complex ion:</p>
      
  <ul>
    <p ALIGN="left">1. <b>Name the ligands first, in alphabetical order, then 
      the metal atom or ion</b>. <b>Note: </b>The <b>metal atom or ion</b> is 
      written <b>before</b> the <b>ligands</b> in the <b>chemical formula</b>. 
    </p>
    <p ALIGN="left">2. The names of some common ligands are listed in Table 1.</p>
    <ul>
      <p ALIGN="left"><i><font FACE="Symbol">� </font>For anionic ligands end 
        in &quot;-o&quot;; for anions that end in &quot;-ide&quot;(e.g. chloride), 
        &quot;-ate&quot; (e.g. sulfate, nitrate), and &quot;-ite&quot; (e.g. nirite), 
        change the endings as follows: </i><i>-ide <font FACE="Symbol"><img src="arrow.jpg" width="25" height="8"></font> 
        -o; -ate <font FACE="Symbol"><img src="arrow.jpg" width="25" height="8"></font> 
        -ato; -ite <img src="arrow.jpg" width="25" height="8"> -ito </i></p>
      <p ALIGN="left"><i><font FACE="Symbol">� </font>For neutral ligands, the 
        common name of the molecule is used e.g. H<sub>2</sub>NCH<sub>2</sub>CH<sub>2</sub>NH<sub>2</sub> 
        (ethylenediamine). <b>Important exceptions</b>: water is called &#145;aqua&#146;, 
        ammonia is called &#145;ammine&#146;, carbon monoxide is called &#145;carbonyl&#146;, 
        and the N<sub>2</sub> and O<sub>2 </sub>are called &#145;dinitrogen&#146; 
        and &#145;dioxygen&#146;.</i></p>
    </ul>
  </ul>
            </blockquote>

 <blockquote>         
<p ALIGN="left"><b>4</b>. After naming the ligands, name the central metal. If 
  the complex ion is a cation, the metal is named same as the element. For example, 
  Co in a complex cation is call cobalt and Pt is called platinum. (See examples 
  1-4). If the complex ion is an anion, the name of the metal ends with the suffix 
  &#150;ate. (See examples 5 and 6.). For example, Co in a complex anion is called 
  cobaltate and Pt is called platinate. For some metals, the Latin names are used 
  in the complex anions e.g. Fe is called ferrate (not ironate).</p></blockquote>

  <blockquote>       
<p ALIGN="left"><b>5.</b> Following the name of the metal, the oxidation state 
  of the metal in the complex is given as a Roman numeral in parentheses. </p>
       <ul> 
            
    <p ALIGN="left"><b>C</b>. To name a neutral complex molecule, follow the rules 
      of naming a complex cation. <b>Remember:</b> Name the (possibly complex) 
      cation <b>BEFORE </b>the (possibly complex) anion.See examples 7 and 8.</p>
    <p ALIGN="left">For historic reasons, some coordination compounds are called by 
  their common names. For example, Fe(CN)<sub>6</sub><sup>3<font FACE="Symbol">-</font> 
  </sup>and Fe(CN)<sub>6</sub><sup>4<font FACE="Symbol">-</font> </sup>are named 
  ferricyanide and ferrocyanide respectively, and Fe(CO)<sub>5</sub> is called 
  iron carbonyl.</ul></blockquote>
<p ALIGN="left"> <b><u>Examples</u> Give the systematic names for the following 
  coordination compounds:</b></p>


  
    <p ALIGN="left">1. [Cr(NH<sub>3</sub>)<sub>3</sub>(H<sub>2</sub>O)<sub>3</sub>]Cl<sub>3</sub> 
    </p>
  <ul>
  <p ALIGN="left"><font color="#0000ff"><i><b>Answer: triamminetriaquachromium(III) 
    chloride</b></i></font></p>
      
      
        
          
            <p ALIGN="left"><font color="#0000ff"><i><b>Solution: The complex 
              ion is inside the parentheses, which is a cation.</b></i></font></p>
            <p ALIGN="left"><font color="#0000ff"><i><b>The ammine ligands are 
              named before the aqua ligands according to alphabetical order.</b></i></font></p>
            <p ALIGN="left"><font color="#0000ff"><i><b>Since there are three 
              chlorides binding with the complex ion, the charge on the complex 
              ion must be +3 ( since the compound is electrically neutral).</b></i></font></p>
            <p ALIGN="left"><font color="#0000ff"><i><b>From the charge on the 
              complex ion and the charge on the ligands, we can calculate the 
              oxidation number of the metal. In this example, all the ligands 
              are neutral molecules. Therefore, the oxidation number of chromium 
              must be same as the charge of the complex ion, +3.</b></i></font></ul>
          
        
      
    
    <p ALIGN="left">2. [Pt(NH<sub>3</sub>)<sub>5</sub>Cl]Br<sub>3</sub> </p>
  
  <ul><p ALIGN="left"><font color="#0000ff"><i><b>Answer: pentaamminechloroplatinum(IV) 
    bromide</b></i></font></p>
  
  
    <p ALIGN="left"><font color="#0000ff"><i><b>Solution: The complex ion is a 
      cation, the counter anion is the 3 bromides.</b></i></font></p>
   
            <p ALIGN="left"><font color="#0000ff"><i><b>The charge of the complex 
              ion must be +3 since it bonds with 3 bromides.</b></i></font></p>
            <p ALIGN="left"><font color="#0000ff"><i><b>The NH<sub>3</sub> are 
              neutral molecules while the chloride carries <font
            FACE="Symbol">-</font> 1 charge. Therefore, the oxidation number of 
              platinum must be +4. </b></i></font></ul>
          
        
      
    
    <p ALIGN="left">3. [Pt(H<sub>2</sub>NCH<sub>2</sub>CH<sub>2</sub>NH<sub>2</sub>)<sub>2</sub>Cl<sub>2</sub>]Cl<sub>2</sub> 
    </p>
  
  
<blockquote>
  <p><font color="#0000ff"><i><b>Answer: dichlorobis(ethylenediamine)platinum(IV) 
    chloride</b></i></font></p>
  <p><font color="#0000ff"><i><b>Solution: ethylenediamine is a bidentate ligand, 
    the bis- prefix is used instead of di-</b></i></font></p>
</blockquote>
<p ALIGN="left">4. [Co(H<sub>2</sub>NCH<sub>2</sub>CH<sub>2</sub>NH<sub>2</sub>)<sub>3</sub>]<sub>2</sub>(SO<sub>4</sub>)<sub>3</sub> 
    </p>
  
  
<blockquote>
  <p><font color="#0000ff"><i><b>Answer: tris(ethylenediamine)cobalt(III) sulfate</b></i></font></p>
  <p><font color="#0000ff"><i><b>Solution: The sulfate is the counter anion in 
    this molecule. Since it takes 3 sulfates to bond with two complex cations, 
    the charge on each complex cation must be +3.</b></i></font></p>
  <p><font color="#0000ff"><i><b>Since ethylenediamine is a neutral molecule, 
    the oxidation number of cobalt in the complex ion must be +3.</b></i></font></p>
  <p><font color="#0000ff"><i><b>Again, remember that you never have to indicate 
    the number of cations and anions in the name of an ionic compound.</b></i></font></p>
</blockquote>
<p ALIGN="left">5. K<sub>4</sub>[Fe(CN)<sub>6</sub>] </p>
  
  
<blockquote>
  <p><font color="#0000ff"><i><b>Answer: potassium hexacyanoferrate(II) </b></i></font></p>
  <p><font color="#0000ff"><i><b>Solution: potassium is the cation and the complex 
    ion is the anion.</b></i></font></p>
  <p><font color="#0000ff"><i><b>Since there are 4 K<sup>+</sup> binding with 
    a complex ion, the charge on the complex ion must be <font FACE="Symbol">-</font> 
    4.</b></i></font></p>
  <p><font color="#0000ff"><i><b>Since each ligand carries &#150;1 charge, the 
    oxidation number of Fe must be +2.</b></i></font></p>
  <p><font color="#0000ff"><i><b>The common name of this compound is potassium 
    ferrocyanide.</b></i></font></p>
</blockquote>
<p ALIGN="left">6. Na<sub>2</sub>[NiCl<sub>4</sub>] </p>
  
  
<blockquote>
  <p><font color="#0000ff"><i><b>Answer: sodium tetrachloronickelate(II)</b></i></font></p>
  <p><font color="#0000ff"><i><b>Solution: The complex ion is the anion so we 
    have to add the suffix &#150;ate in the name of the metal.</b></i></font></p>
</blockquote>
<p ALIGN="left">7. Pt(NH<sub>3</sub>)<sub>2</sub>Cl<sub>4</sub> </p>
  
  
<blockquote>
  <p><font color="#0000ff"><i><b>Answer: diamminetetrachloroplatinum(IV)</b></i></font></p>
  <p><font color="#0000ff"><i><b>Solution: This is a neutral molecule because 
    the charge on Pt<sup>+4</sup> equals the negative charges on the four chloro 
    ligands.</b></i></font></p>
  <p><font color="#0000ff"><i><b>If the compound is [Pt(NH<sub>3</sub>)<sub>2</sub>Cl<sub>2</sub>]Cl<sub>2</sub>, 
    eventhough the number of ions and atoms in the molecule are identical to the 
    example, it should be named: diamminedichloroplatinum(II) chloride, a big 
    difference.</b></i></font></p>
</blockquote>
<p ALIGN="left">8. Fe(CO)<sub>5</sub> </p>
  
  
<blockquote>
  <p><font color="#0000ff"><i><b>Answer: pentacarbonyliron(0)</b></i></font></p>
  <p><font color="#0000ff"><i><b>Solution: Since it is a neutral complex, it is 
    named in the same way as a complex cation. The common name of this compound, 
    iron carbonyl, is used more often.</b></i></font></p>
</blockquote>
<p ALIGN="left">9. (NH<sub>4</sub>)<sub>2</sub>[Ni(C<sub>2</sub>O<sub>4</sub>)<sub>2</sub>(H<sub>2</sub>O)<sub>2</sub>] 
    </p>
  
  
<blockquote>
  <p><font color="#0000ff"><i><b>Answer: ammonium diaquabis(oxalato)nickelate(II)</b></i></font></p>
  <p><font color="#0000ff"><i><b>Solution: The oxalate ion is a bidentate ligand.</b></i></font></p>
</blockquote>
<p ALIGN="left">10. [Ag(NH<sub>3</sub>)<sub>2</sub>][Ag(CN)<sub>2</sub>] </p>
  
  
<blockquote>
  <p><font color="#0000ff"><i><b>Answer: diamminesilver(I) dicyanoargentate(I)</b></i></font></p>
  <p><font color="#0000ff"><i><b>You can have a compound where both the cation 
    and the anion are complex ions. Notice how the name of the metal differs even 
    though they are the same metal ions.</b></i></font></p>
</blockquote>

  <p><font color="#0000ff"><i><b>2. (NH<sub>4</sub>)<sub>2</sub>[CuCl<sub>4</sub>]</b></i></font></p>
  <p><font color="#0000ff"><i><b>3. Na<sub>3</sub>[FeCl<sub>1</sub>(CN)<sub>5</sub>]</b></i></font></p>
