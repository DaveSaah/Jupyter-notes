```python
import Now

items = {'coke': 1.0, 'fanta': 3.0, 'sprite': 4.0}

itm = []
prc = []
qty = []
atm = []
amt = 0

print('If no item left, Enter \'done\'\n')

while True:
    buy = input('Item bought :').lower()
    if buy == 'done':
        break
    
    itm.append(buy)
    price = items[buy]
    prc.append(price)
    quant = int(input('How many {0} were bought: '.format(buy)))
    qty.append(quant)
    atm.append(quant * price)
    amt += amt + (quant * price)

print('\nThe amount is {0}'.format(amt))
paid = float(input('How much was paid: '))
bal = paid - amt
print('\nBalance is: {}'.format(bal))

print('\n\n\n\n\n')
Now.timestamp()
print('\n')
print("{0:^40}".format('SUMMARY'))
print('_'*40)
print("{0:<10}{1:<10}{2:>10}{3:>10}".format('Items', 'Price', 'Quantity', 'Amount'))
print('_'*40)

for x in range(len(itm)):
    print("{0:<10}{1:<10}{2:>10}{3:>10}".format(itm[x], prc[x], qty[x], atm[x]))


print('{0:>40}'.format('_'*40))
print('{0:>30}{1:>10}'.format('Total Amount', amt))
print('{0:>40}'.format('_'*40))
print('\nPaid: {0}\nBalance: {1}'.format(paid, bal))
```
