```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
%pylab inline

x = np.linspace(-10, 10, 1000)
y = x

#x0 = linspace(0, 0)
#y0 = linspace(-10, 10)

#x1 = linspace(-10, 10)
#y1 = linspace(0, 0)

plt.title('A graph of y = x')
plt.plot(x, y)
#plt.plot(x0, y0, color='red')
#plt.plot(x1, y1, color='red')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.grid()

plt.show()
```
